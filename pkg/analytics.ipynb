```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

```

```python
df = pd.DataFrame({
    'Feature1': [0, 1, 2, 3, 4],
    'Feature2': [10, 12, 15, 14, 13],
    'Feature3': [5, 7, 6, 5, 8],
    'Feature4': [1, 4, 3, 2, 4],
    'sensor4': [20, 21, 19, 22, 20],
})
```

```python
plot_config = [
    {'x' : 'Feature1', 'y' : 'Feature2'},
    {'x' : 'Feature1', 'y' : 'Feature3'},
    {'x' : 'Feature1', 'y' : 'Feature4'},
    {'x' : 'Feature2', 'y' : 'Feature3'},
    {'x' : 'Feature2', 'y' : 'Feature4'},
    {'x' : 'Feature3', 'y' : 'Feature4'},
]


num_plots = len(plot_config)
cols = 2
rows = math.ceil(num_plots / cols)

fig, axes = plt.subplots(rows, cols, figsize=(10, 5 * rows))
axes = axes.flatten() if num_plots > 1 else [axes]
axes = np.array(axes).flatten()

for i, plot_cfg in enumerate(plot_config):
    x_data = df[plot_cfg['x']]
    y_data = df[plot_cfg['y']]
    r = np.corrcoef(x_data, y_data)
    axes[i].grid(True, linestyle='--', linewidth=0.5)
    axes[i].scatter(x_data, y_data)
    axes[i].set_title(f"Correlation = {round(r[1,0], 2)}")
    axes[i].set_xlabel(plot_cfg['x'])
    axes[i].set_ylabel(plot_cfg['y'])
    


# Hide unused subplots
for j in range(num_plots, len(axes)):
    fig.delaxes(axes[j])



plt.tight_layout()
plt.show()
```
